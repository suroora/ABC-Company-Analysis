## Step:2 - Analysis Tasks:

##### Q1. Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees.

```python
team_count = da["Team"].value_counts().sort_values(ascending=False) 
team_prec=(team_count  / len(da) * 100).round(2)
team_df = pd.DataFrame({"Count": team_count,"precentage": team_prec})
team_df
```

```python
#visualization of the team distribution using a bar chart
plt.Figure()
team_count.plot(kind='bar')
plt.title("Employee per Team")
plt.xlabel("Team"); plt.ylabel("Count")
plt.ylim(12, 20)
plt.tight_layout()
plt.show()
```

```python
#Save Distributrion data to a csv file
team_df.to_csv(r"1 - Learning Path\ABC company analysis\ABC-Company-Analysis\Team_distribution.csv", index=True)
```


##### Q2. Segregate employees based on their positions within the company.

```python
#Segregate employees based on their positions within the company.
position_count = da["Position"].value_counts().sort_values(ascending=False)
position_count
```

```python
# visualize the employee position distribution using a pie chart
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
plt.suptitle("Employee Position Distribution")
position_count.plot(kind='pie', autopct='%1.1f%%', startangle=140, ax=axes[0])
position_count.plot(kind="bar", ax=axes[1])
plt.tight_layout()
plt.show()
```

##### Q3. Identify the predominant age group among employees.

```python
# predominant age group among employees.

da["Age"].describe()
```

```python
# predominant age group among employees.
age_bins = [10, 20, 30, 40,50]
age_labels = ['10-19','20-29', '30-39', '40-49']
da['Age_Group'] = pd.cut(da['Age'], bins=age_bins, labels=age_labels, right=False)
age_group_count = da['Age_Group'].value_counts().sort_index()  
age_group_count 
```

```python
#visualization of the age group distribution using a bar chart
plt.figure()
age_group_count.plot(kind="bar")
plt.title("Employee Count by Age Group")
plt.xlabel("Age Group"); plt.ylabel("Count")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

##### Q4. Discover which team and position have the highest salary expenditure.

```python
# Discover which team and position have the highest salary expenditure.
salary_expenditure = da.groupby(['Team', 'Position'])['Salary'].sum().reset_index()
highest_expenditure = salary_expenditure.loc[salary_expenditure['Salary'].idxmax()]    
highest_expenditure
```

```python
salary_by_team = (
    df.groupby("Team", dropna=False)["Salary"]
      .sum(min_count=1)
      .fillna(0)
      .sort_values(ascending=False)
)
salary_by_position = (
    df.groupby("Position", dropna=False)["Salary"]
      .sum(min_count=1)
      .fillna(0)
      .sort_values(ascending=False)
)

print("Top-spend Team:", salary_by_team.index[0], int(salary_by_team.iloc[0]))
print("Top-spend Position:", salary_by_position.index[0], int(salary_by_position.iloc[0]))
```

```python

fig, axes = plt.subplots(1, 2, figsize=(12, 6))
salary_by_team.plot(kind="bar", ax=axes[0], color='skyblue', title='Total Salary by Team')
salary_by_position.plot(kind="bar", ax=axes[1], color='salmon', title='Total Salary by Position')
plt.tight_layout()  
plt.show()
```

##### Q5. Investigate if there's any correlation between age and salary, and represent it visually.

```python
#Investigate if there's any correlation between age and salary, and represent it visually.
correlation = da[['Age', 'Salary']].dropna().corr()
correlation
```

```python
#visualization correlation between age and salary

plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation between Age and Salary") 
plt.show()
```

```python
age_salary = df[["Age", "Salary"]].dropna()
corr = age_salary["Age"].corr(age_salary["Salary"])
print("Pearson r (Age vs Salary):", round(corr, 3))

plt.figure()
plt.scatter(age_salary["Age"], age_salary["Salary"])
# Trendline
z = np.polyfit(age_salary["Age"], age_salary["Salary"], 1)
p = np.poly1d(z)
xs = np.linspace(age_salary["Age"].min(), age_salary["Age"].max(), 100)
plt.plot(xs, p(xs))
plt.title(f"Age vs Salary (r = {corr:.3f})")
plt.xlabel("Age"); plt.ylabel("Salary")
plt.tight_layout()
plt.show()
```
